```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
mix_h = pd.read_csv('~/cbs-covid/src/data/Mix_h.csv',header=None)
mix_o = pd.read_csv('~/cbs-covid/src/data/Mix_o.csv',header=None)
mix_s = pd.read_csv('~/cbs-covid/src/data/Mix_s.csv',header=None)
mix_w = pd.read_csv('~/cbs-covid/src/data/Mix_w.csv',header=None)
```

```python
%matplotlib widget
fig, axs = plt.subplots(2, 2)

all_matrices = [mix_h, mix_o, mix_s, mix_w]
all_matrices_name = ["home", "other", "school", "work"]

i = 0
for ax in axs.flat:
    im = ax.imshow(all_matrices[i], cmap='hot', interpolation='nearest')
    ax.set_title(all_matrices_name[i])
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im, cax=cax)
    i += 1
plt.show()
```
